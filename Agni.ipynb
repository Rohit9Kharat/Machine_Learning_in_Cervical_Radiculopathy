```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_sneh = pd.read_excel('For Statistics.xlsx', sheet_name='DataSnehpana', header=0)
df_basti = pd.read_excel('For Statistics.xlsx', sheet_name='DataBasti', header=0)
df_abhya = pd.read_excel('For Statistics.xlsx', sheet_name='DataAbhyanga', header=0)
```

```python
sneh_manda = df_sneh[df_sneh['Agni']=='Manda'].describe().T
sneh_tikshna = df_sneh[df_sneh['Agni']=='Tikshna'].describe().T
sneh_visham = df_sneh[df_sneh['Agni']=='Visham'].describe().T

basti_manda = df_basti[df_basti['Agni']=='Manda'].describe().T
basti_tikshna = df_basti[df_basti['Agni']=='Tikshna'].describe().T
basti_visham = df_basti[df_basti['Agni']=='Visham'].describe().T

abhya_manda = df_abhya[df_abhya['Agni']=='Manda'].describe().T
abhya_tikshna = df_abhya[df_abhya['Agni']=='Tikshna'].describe().T
abhya_visham = df_abhya[df_abhya['Agni']=='Visham'].describe().T
```

```python
sneh_manda.to_excel('H:/DATASETS/Pinky Maushi/Agni/sneh_manda_statistics.xlsx')
sneh_tikshna.to_excel('H:/DATASETS/Pinky Maushi/Agni/sneh_tikshna_statistics.xlsx')
sneh_visham.to_excel('H:/DATASETS/Pinky Maushi/Agni/sneh_visham_statistics.xlsx')

basti_manda.to_excel('H:/DATASETS/Pinky Maushi/Agni/basti_manda_statistics.xlsx')
basti_tikshna.to_excel('H:/DATASETS/Pinky Maushi/Agni/basti_tikshna_statistics.xlsx')
basti_visham.to_excel('H:/DATASETS/Pinky Maushi/Agni/basti_visham_statistics.xlsx')

abhya_manda.to_excel('H:/DATASETS/Pinky Maushi/Agni/abhya_manda_statistics.xlsx')
abhya_tikshna.to_excel('H:/DATASETS/Pinky Maushi/Agni/abhya_tikshna_statistics.xlsx')
abhya_visham.to_excel('H:/DATASETS/Pinky Maushi/Agni/abhya_visham_statistics.xlsx')
```

## Manda

```python
s_manda = sneh_manda[sneh_manda.index != 'Age']['mean'].to_frame().T
a_manda = abhya_manda[abhya_manda.index != 'Age']['mean'].to_frame().T
b_manda = basti_manda[basti_manda.index != 'Age']['mean'].to_frame().T

frames = [s_manda, a_manda, b_manda]

result_manda = pd.concat(frames)
```

```python
result_manda.index = ['Snehpana', 'Abhyanga', 'Matrabasti']

result_manda
```

```python
result_manda.to_excel('H:/DATASETS/Pinky Maushi/Agni/result_manda.xlsx')
#result_manda.to_csv('H:/DATASETS/Pinky Maushi/Agni/result_manda.csv')
```
